```python
IQR = X_train['radius_mean'].quantile(0.75)-X_train['radius_mean'].quantile(0.25)
Lower_fence = X_train['radius_mean'].quantile(0.25) - (IQR*3)
Upper_fence = X_train['radius_mean'].quantile(0.75) + (IQR*3)
print('radius_mean outliers are values < {lowerboundary} or > {upperboundary}'.format(lowerboundary=Lower_fence, upperboundary=Upper_fence))
```

```python
IQR = X_train['concavity_mean'].quantile(0.75)-X_train['concavity_mean'].quantile(0.25)
Lower_fence = X_train['concavity_mean'].quantile(0.25) - (IQR*3)
Upper_fence = X_train['concavity_mean'].quantile(0.75) + (IQR*3)
print('concavity_mean outliers are values < {lowerboundary} or > {upperboundary}'.format(lowerboundary=Lower_fence, upperboundary=Upper_fence))
```
